```python
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
df = pd.read_csv("/home/yushi/code/churn_broadcasting/data/preprocessed/after_eda_data.csv")

df.head()
```

```python
# Ordinal Encoding for the Contract feature
df["Contract"] = df["Contract"].astype("category").cat.codes
```

```python
# One-hot encoding for the InternetService feature
one_hot_encoding = pd.get_dummies(df['InternetService']).astype(int)

df = pd.concat([df, one_hot_encoding], axis=1)

df = df.drop('InternetService', axis=1)
```

```python
df
```

```python
# We will use a Decision tree model to predict the output so we don't need to scale the continuous features
# Now we save the new processed data
df.to_csv("/home/yushi/code/churn_broadcasting/data/preprocessed/processed_xgboost_data.csv")
```

```python
# Now we can do some preprocessing for the logistic regression model, so we scale the continuous features
scaler = sk.preprocessing.RobustScaler()

df[['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(df[['tenure', 'MonthlyCharges', 'TotalCharges']])

df
```
